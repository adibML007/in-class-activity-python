```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
pd.options.display.max_rows = 999
warnings.filterwarnings("ignore")
```

```python
#loading the data set
link = "./Data/CySecData.csv"
df = pd.read_csv(link)
```

```python
df.head()
```

```python
#Info command
df.info()
```

```python
#Creating the dummy values for the object data types
dfDummies = pd.get_dummies(df,columns=['protocol_type','service','flag'])
```

```python

dfDummies = dfDummies.drop(['class'], axis=1)
```

```python
#Importing the standard Scaler
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler_df = scaler.fit_transform(dfDummies)
```

```python
dfNormalized = pd.DataFrame(scaler_df, columns = dfDummies.columns)
```

```python
dfNormalized
```

```python
X = dfNormalized
y = df['class']
```

```python
#Importing the required libraries for the models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('SVM', SVC()))
models.append(('RandomForestClassifier', RandomForestClassifier()))
```

```python
#Calculating the accuracy for each fold and diplaying the mean and STD
from numpy import mean
from numpy import std
for name, model in models:
    kfold = KFold(n_splits=10, random_state=7, shuffle=True)
    cv_results = cross_val_score(model, X, y, cv=kfold, scoring='accuracy')
    print(f'Model:{name}, accurancy:{cv_results}, mean: {mean(cv_results)}, std: {std(cv_results)}')
```
